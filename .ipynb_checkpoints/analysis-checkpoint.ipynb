### 2. Merge with content moderation table ###

```python
label_to_taxonomy = {
    'Hate/Identity Hate': 'Hate Speech',
    'Sexual': 'Sexual Content',
    'Sexual (minor)': 'Sexual Content',
    'Violence': 'Violence',
    'Suicide and Self Harm': 'Self-Harm',
    'Threat': 'Threats',
    'Guns and Illegal Weapons': 'Other Risks',
    'Controlled/Regulated Substances': 'Other Risks',
    'Criminal Planning/Confessions': 'Criminal Planning',
    'PII/Privacy': 'Privacy Violations',
    'Harassment': 'Harassment',
    'Profanity': 'Profanity',
    'Other': 'Other Risks',
    'Needs Caution': 'Other Risks',
    'Safe': 'Other Risks',
    'Deception/Fraud': 'Other Risks'
}
```

```python
file_path = '/Users/luciadelatorre/Desktop/COLUMBIA/GitHub/trust-safety-ai-analysis/datasets/aegis_train.parquet'
aegis_train = pd.read_parquet(file_path)

taxonomy = [
    {"Category": "Hate Speech", "Description": "Language targeting groups based on identity"},
    {"Category": "Hallucinations", "Description": "Model-generated false or misleading information"},
    {"Category": "Privacy Violations", "Description": "Exposure of PII or sensitive data"},
    {"Category": "Bias", "Description": "Language reflecting or amplifying harmful stereotypes"},
    {"Category": "Violence", "Description": "Content promoting or describing physical harm"},
    {"Category": "Sexual Content", "Description": "Sexually explicit or suggestive content"},
    {"Category": "Harassment", "Description": "Targeted abuse or intimidation"},
    {"Category": "Self-Harm", "Description": "Content promoting or describing self-injury"},
    {"Category": "Criminal Planning", "Description": "Instructions or encouragement for illegal acts"},
    {"Category": "Profanity", "Description": "Use of offensive language"},
    {"Category": "Threats", "Description": "Content expressing intent to harm"},
    {"Category": "Misinformation", "Description": "False or misleading information"},
    {"Category": "Other Risks", "Description": "Other safety or policy concerns"}
]
taxonomies = pd.DataFrame(taxonomy)

label_to_taxonomy = {
    'Hate/Identity Hate': 'Hate Speech',
    'Sexual': 'Sexual Content',
    'Sexual (minor)': 'Sexual Content',
    'Violence': 'Violence',
    'Suicide and Self Harm': 'Self-Harm',
    'Threat': 'Threats',
    'Guns and Illegal Weapons': 'Other Risks',
    'Controlled/Regulated Substances': 'Other Risks',
    'Criminal Planning/Confessions': 'Criminal Planning',
    'PII/Privacy': 'Privacy Violations',
    'Harassment': 'Harassment',
    'Profanity': 'Profanity',
    'Other': 'Other Risks',
    'Needs Caution': 'Other Risks',
    'Safe': 'Other Risks',
    'Deception/Fraud': 'Other Risks'
}

def map_labels(label_str):
    if pd.isna(label_str):
        return []
    labels = [l.strip() for l in label_str.split(',')]
    mapped = [label_to_taxonomy.get(l, 'Other Risks') for l in labels]
    return list(set(mapped))  # remove duplicates

aegis_train['taxonomy_categories'] = aegis_train['labels_0'].apply(map_labels)

exploded = aegis_train.explode('taxonomy_categories')

merged = exploded.merge(taxonomies, left_on='taxonomy_categories', right_on='Category', how='left')

print(merged[['id', 'text', 'labels_0', 'taxonomy_categories', 'Description']].head())
```

### 3. Create frequency of risks by category table ###

```python
pivot = merged.pivot_table(index='taxonomy_categories', columns='text_type', values='id', aggfunc='count').fillna(0)
pivot.to_csv('datasets/category_by_type.csv')
```

### 4. Create risk by source table ###

```python
pivot = merged.pivot_table(index='taxonomy_categories', columns='text_type', values='id', aggfunc='count', fill_value=0)
pivot.reset_index().to_csv('risk_by_source.csv', index=False)
```

### 5. Create risk co-ocurrence table and heatmap ###

```python
from itertools import combinations
import numpy as np

grouped = merged.groupby('id')['taxonomy_categories'].apply(list)

categories = sorted(merged['taxonomy_categories'].dropna().unique())
matrix = pd.DataFrame(0, index=categories, columns=categories)

for cats in grouped:
    for a, b in combinations(set(cats), 2):
        matrix.loc[a, b] += 1
        matrix.loc[b, a] += 1
    for c in set(cats):
        matrix.loc[c, c] += 1

matrix.to_csv('risk_cooccurrence.csv')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/Users/luciadelatorre/Desktop/COLUMBIA/GitHub/trust-safety-ai-analysis/datasets/risk_cooccurrence.csv'

co_occurrence_df = pd.read_csv(file_path, index_col=0)

plt.figure(figsize=(10, 8))
sns.heatmap(
    co_occurrence_df,
    annot=True,
    fmt='d',
    cmap='coolwarm',
    linewidths=0.5,
    vmin=0,
    vmax=co_occurrence_df.values.max() * 0.1
)
plt.title('Co-occurrence of Risk Categories in LLM Conversations')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### 6. Create ambiguous risks table ###

```python
ambiguous = merged['taxonomy_categories'] == 'Other Risks'
ambiguous_counts = pd.Series({
    'Ambiguous/Other': ambiguous.sum(),
    'Standard Category': (~ambiguous).sum()
}).reset_index()
ambiguous_counts.columns = ['Type', 'Count']
ambiguous_counts.to_csv('ambiguous_risks.csv', index=False)
```

### 7. Create multi label complexity table ###

```python
multi_label_counts = merged.groupby('id')['taxonomy_categories'].nunique().value_counts().sort_index()

multi_label_df = multi_label_counts.reset_index()
multi_label_df.columns = ['Number of Risk Categories', 'Number of Conversations']

multi_label_df.to_csv('multi_label_complexity.csv', index=False)
```
